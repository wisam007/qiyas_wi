```python
import pandas as pd
import numpy as np

data = {
'Patient_ID':['P001','P002','P003','P004','P004'],
'Full_Name':['Abebe Kebede',' Hana Tesfaye ','Samuel Bekele',

'Meron Alemu','Meron Alemu'],

'Age':[35,np.nan,42,29,29],
'Address':[
'Bole Road, Addis Ababa, Bole, 03, H125',
'CMC Road, Addis Ababa, Yeka, 10, H201',
'Piassa, Addis Ababa, Arada, 01, H505',
'Megenagna, Addis Ababa, Bole, 05, H901',
'Megenagna, Addis Ababa, Bole, 05, H901'

]
}

df = pd.DataFrame(data)
```

```python
# 1. Display the dataset.
print(df)
```

```python
# 2. Find the number of missing values in each column.
print(df.isna())
# 3. Replace missing ages with the mean age.
df['Age'] = df['Age'].fillna(df['Age'].mean())

print(df)
```

```python
 # 4. Remove duplicate rows.
print(df.duplicated())
df = df.drop_duplicates(subset=["Full_Name"])
print(df)
```

```python
# 5. Remove leading and trailing spaces from Full_Name
df["Full_Name"] = df["Full_Name"].str.strip()

print(df["Full_Name"])
```

```python

# 6. Split Full_Name into First_Name and Last_Name.
df[["First_Name", "Last_Name"]] = df["Full_Name"].str.split(" ",expand=True)
print(df)

```

```python
# 7. Split Address into Street, City, Subcity, Woreda, and House_No.
df[["Street", "City", "Subcity", "Woreda","House_No."]] = df["Address"].str.split(",",expand=True)

# 8. Display the cleaned dataset.
print(df)
```

```python
"""Exercise 2: Online Store Sales Report ★★"""

data = {
'Month':['Jan','Jan','Feb','Feb','Mar','Mar'],
'Product':['Laptop','Phone','Laptop','Phone','Laptop','Phone'],
'Sales':[15,np.nan,18,20,15,25]
}

df = pd.DataFrame(data)
```

```python
# 1. Display the dataset.
print(df)
# 2. Identify missing values.
print(df.isna())
# 3. Replace missing sales with zero.
df["Sales"] = df["Sales"].fillna(0)
print(df)
```

```python
 #4. Calculate total sales.

total_sales = df["Sales"].sum()
print("Total Sales:", total_sales)
print("")
# 5. Create a pivot table:


# Rows → Month
# Columns → Product
# Values → Sales

df_pivot = df.pivot_table(
    index="Month",
    columns="Product",
    values="Sales"
)
print(df_pivot)
# 6. Replace missing values in the pivot table with zero.
df_pivot_filled = df_pivot.fillna(0)
# 7. Find the month with the highest total sales.
df_sales_sum = df_pivot.sum(axis=1)

print(f"The month with the higest sales is '{df_sales_sum.idxmax()}' with {df_sales_sum.max()}")
```

```python
"""Exercise 3: Employee Email Processing
★★
Dataset"""

data = {
'Employee_ID':['E1','E2','E3','E4'],
'Email':[
'abebe@gmail.com',
'hana@yahoo.com',
'samuel@outlook.com',
'meron@gmail.com'
]
}

df = pd.DataFrame(data)
```

```python
# 1. Split Email into Username and Domain.
df[["Username", "Domain"]] = df["Email"].str.split("@" , expand=True)
# 2. Convert usernames to uppercase.
df["Username"] = df["Username"].str.upper()
print(df)
# 3. Create a Company_Email column ending with "@company.com".
df["Company_Email"] = df["Username"].str.lower() + "@company.com"
print(df)
# 4. Count the number of employees using Gmail.
is_gmail = df["Email"].str.contains("gmail")
print("Number of employees using Gmail:-",is_gmail.sum())
 # 5. Display only employees with Gmail accounts.
print(df[is_gmail])
```
