```python
"""
Challenge Exercise: Customer Orders
Analysis ★★★★★

Dataset"""

import numpy as np
import pandas as pd

data = {
'Customer_ID':['C001','C002','C003','C003'],
'Full_Name':['Abebe Kebede',
' Hana Alemu ',
'Meron Tadesse',
'Meron Tadesse'],

'Address':[
'Bole Road, Addis Ababa, Bole, 03, H125',
'CMC Road, Addis Ababa, Yeka, 10, H201',
'Piassa, Addis Ababa, Arada, 01, H505',
'Piassa, Addis Ababa, Arada, 01, H505'
],
'Product':['Laptop','Phone','Laptop','Tablet'],
'Amount':[45000,np.nan,32000,28000]
}

df = pd.DataFrame(data)

# Tasks
# 1. Find missing values.
print(df.isna().sum())
# 2. Replace missing Amount values with the average amount.
avg_amt = df['Amount'].mean()
df['Amount'] = df['Amount'].fillna(avg_amt)
# 3. Remove duplicate customers.
df_clean = df.drop_duplicates().reset_index(drop=True)
print(df_clean)
# 4. Remove leading and trailing spaces from Full_Name.
df_clean['Full_Name'] = df_clean['Full_Name'].str.strip()
# 5. Split Full_Name into First_Name and Last_Name.
df_clean[['First_Name', 'Last_Name']] = df_clean['Full_Name'].str.split(' ', expand=True)
# 6. Split Address into Street, City, Subcity, Woreda, and House_No.

df_clean[['Street', 'City', 'Subcity', 'Woreda', 'House_No']] = df_clean['Address'].str.split(', ', expand=True)
# 7. Create a Count column.
df_clean['Count'] = 1
# 8. Create a pivot table showing:
# Customer_ID as rows.

# Product as columns.
# Count as values.
df_pivot = df_clean.pivot_table(
    index='Customer_ID',
    columns='Product',
    values='Count',
    aggfunc='sum',
    fill_value=0
)
print(df_pivot)
# 9. Find the customer who spent the most money.
customer_spending = df_clean.groupby('Customer_ID')['Amount'].sum()
top_spender = customer_spending.idxmax()
print(f"Customer who spent the most money: {top_spender}")
# 10. Calculate the average purchase amount.
Avarage_Purchase = df_clean['Amount'].mean()
print("Average Purchase: ",Avarage_Purchase)
# 11. Display customers whose purchases are above average.
print("Above average purchaser is: ")
print(df_clean[df_clean['Amount'] > Avarage_Purchase][['Customer_ID', 'Full_Name']])
print("==========================================================")
# 12. Create a Full_Address column by combining the split address columns.
df_clean['Full_Address'] = (df_clean['Street'] + "," + 
                            df_clean['City'] + "," + 
                            df_clean['Subcity'] + "," + 
                            df_clean['Woreda'] + "," + 
                            df_clean['House_No'])

# 13. Remove the intermediate address columns.

df_clean = df_clean.drop(columns=["Address",'Street', 'City', 'Subcity', 'Woreda', 'House_No'])
# 14. Print cleaned dataset
print(df_clean)
```
