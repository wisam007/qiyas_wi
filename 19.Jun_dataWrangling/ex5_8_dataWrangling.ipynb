```python
"""Exercise 5: Telecom Customer Usage
Dashboard ★★★"""

import numpy as np
import pandas as pd

data = {
'Customer_ID':['C1','C1','C2','C2','C1','C3'],
'Month':['Jan','Jan','Jan','Jan','Feb','Feb'],
'Service':['Data','Voice','Data','Voice','Data','Voice'],
'Usage_GB':[12,5,8,10,np.nan,15]
}

df = pd.DataFrame(data)
print(df)
# Tasks
# 1. Find missing values.
missing_data = df.isna().sum()
print(missing_data)
```

```python

# 2. Replace missing Usage_GB values with the average usage.
from pandas import pivot


df["Usage_GB"] = df["Usage_GB"].fillna(df["Usage_GB"].mean())
print(df)
print("=" *50)
# 3. Create a pivot table:
# Rows → Customer_ID
# Columns → Month and Service
# Values → Usage_GB

df_pivot = df.pivot_table(index="Customer_ID",columns=["Month","Service"] , values="Usage_GB")

# 4. Replace missing values with zero.
df_pivot = df_pivot.fillna(0)
print(df_pivot)
# 5. Find the customer with the highest total usage.
highest_total = df_pivot.sum(axis=1)
print("Customer with the higest total usage is: " ,highest_total.idxmax())
# 6. Calculate average usage for each customer.
average_usage = df_pivot.mean(axis=1)
print(average_usage)
```

```python
"""Exercise 6: Product Code Cleaning and
Analysis ★★★
Dataset
"""
from numpy import int16


data = {
'Product_Code':[
'TV-2025-001',
'LAP-2025-017',
'PHN-2024-023',
'TAB-2025-011'
],
'Price':['45000','65000','32000','55000']
}

df = pd.DataFrame(data)

# Tasks
# 1. Convert Price into integer.
df['Price'] = df['Price'].astype(int)
print(df["Price"].dtype)
# 2. Split Product_Code into:
# Category
# Year
# Product_Number
df[["Category","Year","Product_Number"]] = df["Product_Code"].str.split("-",expand=True)
print(df)
# 3. Find the average product price.
print("Average Product Price: ",df["Price"].mean())
# 4. Find the most expensive product.
print(f"The most expensive profuct is: {df.loc[df["Price"].idxmax(),"Product_Code"]} with price of: {df["Price"].max()}")
# 5. Display all products released in 2025
print("==products released in 2025==")
print(df[df["Year"] == "2025"])
# 6. Calculate total sales value assuming one unit of each product is sold.
total_sales = df["Price"].sum()
print("Total Sales: ",total_sales)
```

```python
"""Exercise 7: University Student
Information Cleaning ★★★★
Dataset"""

data = {
'Student_ID':['S1','S2','S3','S3'],
'Full_Name':[
'Abel Bekele',
' Hana Alemu ',
'Meron Tadesse',
'Meron Tadesse'
],
'Department':['CS','IT',None,'IT'],
'CGPA':[3.6,3.9,3.2,3.2]
}

df = pd.DataFrame(data)

# Tasks
# 1. Display the dataset.
print(df)
# 2. Find missing values.
print(df.isna().sum())
# 3. Remove duplicate rows.
df = df.drop_duplicates(subset='Student_ID' , keep='last')
print(df)
# 4. Remove spaces in Full_Name.
df["Full_Name"] = df['Full_Name'].str.strip()
# 5. Split Full_Name into First_Name and Last_Name.
df[["First_Name","Last_Name"]] = df['Full_Name'].str.split(" ",expand=True)
# 6. Replace missing Department values with "Unknown".
df["Department"] = df['Department'].fillna("Unknown")
# 7. Find the student with the highest CGPA.
print(f"The Student with the highest score is: {df.loc[df['CGPA'].idxmax(),"Full_Name"]}")
# 8. Calculate average CGPA.
average_CGPA = df['CGPA'].mean()
print(f"Average CGPA is: {average_CGPA}")
# 9. Display students whose CGPA is above average.
print(df[df["CGPA"] > average_CGPA]["Full_Name"])
```

```python

"""
Exercise 8: Human Resource Attendance
Analysis ★★★★★
Dataset"""

from sre_constants import POSSESSIVE_REPEAT_ONE
from tarfile import data_filter

from numpy.exceptions import AxisError


data = {
'Employee_ID':['E1','E1','E1','E2','E2','E3','E3'],
'Month':['Jan','Jan','Feb','Jan','Feb','Jan','Feb'],
'Status':['Present',
'Absent',
'Present',
'Present',
'Present',
'Absent',
'Absent']

}

df = pd.DataFrame(data)

# Tasks
# 1. Create a Count column.
df["Count"] =  1
print(df)
# 2. Create a pivot table using:
# Employee_ID as rows.
# Month and Status as columns.
df_pivot = df.pivot_table(index="Employee_ID",columns=[ "Month","Status" ] ,values="Count")
# 3. Replace missing values with zero.
df_pivot = df_pivot.fillna(0)
print(df_pivot)
# 4. Calculate total absences for each employee.
absences = df[df["Status"] == "Absent"].groupby("Employee_ID")["Count"].sum()
print(absences)
#or using pivot table
pivot_absent = df_pivot.loc[:,(slice(None),"Absent")].sum(axis=1)
print(pivot_absent)
# 5. Find the employee with the highest number of absences.
print("Employee with the highest number of absence is ",pivot_absent.idxmax())
# 6. Calculate total attendance records for each employee.
print("total attendance record: ")
total_record = df_pivot.sum(axis=1)
print(total_record)
# 7. Compute the absence percentage for each employee.
absence_av = pivot_absent / total_record * 100
print(absence_av.round(2).astype(str) + "%")
# 8. Determine which employee has the best attendance rate.
print("Employee with best attendance: ",pivot_absent.idxmin())
```
